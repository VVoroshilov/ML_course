```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
mvid = pd.read_csv('data/MVID_101001_171001.txt')
sber = pd.read_csv('data/SBER_101001_171001.txt')
```

```python
res = pd.merge(sber, mvid, on='<DATE>')
res.head()
```

```python
x = res['<CLOSE>_x']
y = res['<CLOSE>_y']
```

```python
class hypothesis:
 
    def __init__(self): 
        self.b0 = 0    
        self.b1 = 0       
    
    def coef(self):
        return "y = {:.2f} + {:.2f}x".format(self.b0, self.b1)
    
    def predict(self, X): 
        return self.b0 + self.b1 * X  
    
    def error(self, X, Y):   
        return sum((self.predict(X) - Y)**2) / (2 * len(X))

    def BGD(self, X, Y, alpha=0.000001, accuracy=0.0001, max_steps=5000):
        steps, errors = [], []
        step = 0        
        old_err = self.error(X, Y)
        new_err = self.error(X, Y) - 1
        dJ = 1
        while (dJ > accuracy) and (step < max_steps):
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1            
            old_err = new_err
            new_err = self.error(X, Y)
            dJ = abs(old_err - new_err) 
            step += 1            
            steps.append(step)
            errors.append(new_err)
        return steps, errors
```

```python
hyp = hypothesis()
steps, errors = hyp.BGD(x, y)
J = hyp.error(x, y)
func = hyp.coef()
print("error after gradient descent:", J)
print(func)
```

```python
X0 = np.linspace(60, 180, 100)
Y0 = hyp.predict(X0)
plt.figure(figsize=(12, 9))
plt.scatter(x, y)
plt.plot(X0, Y0, 'r')
plt.xlabel("Sber")
plt.ylabel("Mvideo")
plt.show()
```

```python
plt.figure()
plt.plot(steps, errors, 'g')
plt.show()
```

```python
x = x.to_numpy().reshape((-1, 1))
y = y.to_numpy().reshape((-1, 1))
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
print(reg.score(x, y))
print("Коэффициенты: \n", reg.coef_)
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x)
plt.figure(figsize=(12, 9))
plt.scatter(x, y, color="black")
plt.plot(x, y_pred, color="blue", linewidth=3)
plt.plot(X0, Y0, 'r')
plt.xlabel("Sber")
plt.ylabel("Mvideo")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("Среднеквадратичная ошибка: %.2f" % mean_squared_error(y, y_pred))
print("Коэффициент детерминации: %.2f" % r2_score(y, y_pred))
```
