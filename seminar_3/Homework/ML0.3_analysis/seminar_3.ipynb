### 5. Постройте круговую диаграмму, показывающую распределение сдавших и не сдавших экзамен

```python
passed = round((df_ppoints[df_ppoints > df_min].shape[0])/(df_ppoints.shape[0]), 4)*100
not_passed = round((df_ppoints[df_ppoints < df_min].shape[0])/(df_ppoints.shape[0]), 4)*100

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

data = [passed, not_passed]
labels = ["Сдали", "Не сдали"]


def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.2f}%".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, labels,
          title="Статус экзамена",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Статистика по сдаче экзамена")

plt.show()
```

### 2. Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В 1 … В к Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.

```python
df_short_ans = df["Задания с кратким ответом"]
df_short_ans
```

```python
df_short_ans = pd.DataFrame({task : df_short_ans.str[task] for task in range(short_ans)})
df_short_ans 
```

```python
for task in range(short_ans):
    point_stats = dict(df_short_ans[task].value_counts(normalize=True))
    passed = 0
    for k, v in point_stats.items():
        if k == "-" or k == "0":
             print(f"Задание {task + 1} не выполнено у {round(v*100, 2)}% людей")
        else:
            passed += v
    print(f"Задание {task + 1} выполнено у {round(passed*100, 2)}% людей", end="\n" + "="*40 + "\n")
#         print(f"Результат {k} у {round(v*100, 2)}% людей")
```

### 4. Сделайте анализ по двум школам:

```python
df_short_ans_117 = df["Задания с кратким ответом"][df["№ школы"] == 117]
df_short_ans_148 = df["Задания с кратким ответом"][df["№ школы"] == 148]
```

```python
df_long_ans_117 = df["Задания с развёрнутым ответом"][df["№ школы"] == 117]
df_long_ans_148 = df["Задания с развёрнутым ответом"][df["№ школы"] == 148]
```

#### 1. По всем выполненным заданиям типа В

```python
def b_class(dataframe):
    df_short_ans = pd.DataFrame({task : dataframe.str[task] for task in range(short_ans)})
    result_df = dict()
    for task in range(short_ans):
        point_stats = dict(df_short_ans[task].value_counts(normalize=True))
        passed = 0
        result_df[task + 1] = dict()
        for k, v in point_stats.items():
            if k == "-" or k == "0":
                 result_df[task + 1]["not_passed"] = v
            else:
                passed += v
        result_df[task + 1]["passed"] = passed
    return pd.DataFrame(result_df)
```

```python
short_stats_117 = b_class(df_short_ans_117)
short_stats_117
```

```python
short_stats_148 = b_class(df_short_ans_148)
short_stats_148
```

```python
first_better = list(short_stats_117.loc["passed"][short_stats_117.loc["passed"] > short_stats_148.loc["passed"]].keys().astype(str))
second_better = list(short_stats_117.loc["passed"][short_stats_117.loc["passed"] < short_stats_148.loc["passed"]].keys().astype(str))
print(f"В школе 117 больше учеников, чем в 148,  выполнили задания: {', '.join(first_better)}")
print(f"В школе 148 больше учеников, чем в 117, выполнили задания: {', '.join(second_better)}")
```

#### 2. По заданиям типа С больше 50%

```python
df_long_ans_117 = df["Задания с развёрнутым ответом"][df["№ школы"] == 117]
df_long_ans_148 = df["Задания с развёрнутым ответом"][df["№ школы"] == 148]
```

```python
answer_mask = df_long_ans_117[0][2::4]
answer_mask 
```

```python
answer_mask = pd.DataFrame({task : float(answer_mask[task])/2 for task in range(long_ans)}, index=[0])
answer_mask
```

```python
df_long_ans_117 = df_long_ans_117.str[::4]
df_long_ans_148 = df_long_ans_148.str[::4]
```

```python
df_long_ans_117 = pd.DataFrame({task : df_long_ans_117.str[task] for task in range(long_ans)})
df_long_ans_117 = df_long_ans_117.astype(np.float64)
df_long_ans_117 
```

```python
df_long_ans_148 = pd.DataFrame({task : df_long_ans_148.str[task] for task in range(long_ans)})
df_long_ans_148 = df_long_ans_148.astype(np.float64)
df_long_ans_148
```

```python
for task in range(long_ans):
    school_117 = (df_long_ans_117[df_long_ans_117[task] >= answer_mask[task][0]].shape[0])/df_long_ans_117.shape[0]
    school_148 = (df_long_ans_148[df_long_ans_148[task] >= answer_mask[task][0]].shape[0])/df_long_ans_148.shape[0]
    print(f"Задание с развёрнутым ответом №{task + 1} выполнили больше людей в школе №{117 if school_117 >= school_148 else 148}")
    
```

#### 3. По среднему баллу юношей и девушек

```python
df_boys = df[df["Пол"] == "М"].copy()
df_girls = df[df["Пол"] == "Ж"].copy()

df_boys_117 = df_boys[df_boys["№ школы"] == 117]
df_boys_148 = df_boys[df_boys["№ школы"] == 148]

df_girls_117 = df_girls[df_girls["№ школы"] == 117]
df_girls_148 = df_girls[df_girls["№ школы"] == 148]
```

```python
boys_mean_117 = round(df_boys_117["Балл"].mean(), 2)
boys_mean_148 = round(df_boys_148["Балл"].mean(), 2)

girls_mean_117 = round(df_girls_117["Балл"].mean(), 2)
girls_mean_148 = round(df_girls_148["Балл"].mean(), 2)

print(f"Юноши из школы 117 имеют средний балл: {boys_mean_117}\nЮноши из школы 148 имеют средний балл: {boys_mean_148}\n")
print(f"Девушки из школы 117 имеют средний балл: {girls_mean_117}\nДевушки из школы 148 имеют средний балл: {girls_mean_148}\n")

print(f"Юноши справились лучше в школе {117 if boys_mean_117 >= boys_mean_148 else 148}")
print(f"Девушки справились лучше в школе {117 if girls_mean_117 >= girls_mean_148 else 148}")
```
