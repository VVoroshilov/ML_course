#### Целевая переменная не приведена к одному виду (цена указана в разных валютах)
#### Для работы с целевой переменной приведём её к одной шкале и численному типу данных (избавимся от указания валюты в значении атрибута Price)

```python
df.info()
```

```python
df["currency"].unique()
```

```python
from forex_python.converter import CurrencyRates
exchanger = CurrencyRates()
currency_values = dict()
for currency in df["currency"].unique():
    if currency == 'Â£':
        currency_values[currency] = exchanger.get_rate("GBP", "EUR")
    else:
        currency_values[currency] = exchanger.get_rate(currency, "EUR")
currency_values
```

```python
df["price_EUR"] = df.apply(lambda x: x["Price"] * currency_values[x["currency"]], axis=1)
df
```

### 5. Провести подготовительную обработку данных, включающую следующие шаги:    

#### 5.1 удаление или заполнение отсутствующих значений

```python
df.isna().sum()
```

```python
df = df[df["Length"].notna()]
df = df[df["Type"].notna()]
df = df[df["Width"].notna()]
df = df[df["Location"].notna()]
df.isna().sum()
```

```python
df["Manufacturer"].value_counts()
```

```python
df.loc[df['Manufacturer'].isna(), "Manufacturer"] = random.choices(list(df['Manufacturer'].value_counts().keys()), weights=list(df['Manufacturer'].value_counts()), k=df['Manufacturer'].isna().sum())
df["Manufacturer"].value_counts()
```

```python
df["Material"].value_counts()
```

```python
df.loc[df['Material'].isna(), "Material"] = random.choices(list(df['Material'].value_counts().keys()), weights=list(df['Material'].value_counts()), k=df['Material'].isna().sum())
df["Material"].value_counts()
```

```python
df.isna().sum()
```

```python
df["Location"].value_counts()
```

```python
df["Country"] = df["Location"].str.split(" ", 1, expand=True)[0]
df
```

```python
df["Country"].value_counts()
```

#### 5.2 приведение всех признаков к бинарной либо числовой шкале

```python
df_cpy = df
df_cpy["Boat Type"] = df_cpy["Boat Type"].str.split(",")
df_cpy["Type"] = df_cpy["Type"].str.split(",")
df_cpy = df_cpy.explode("Boat Type")
df_cpy = df_cpy.explode("Type")
df_cpy
```

```python
df_cpy = pd.get_dummies(data = df_cpy, columns=['Boat Type', 'Type'])
df_cpy.head(3)
```

```python
categorical_cols = df_cpy.columns.drop(["Price", "Manufacturer", "Year Built", "Length", "Width",
                 "Location", "Number of views last 7 days",
                 'currency', 'price_EUR', 'Country',
                ])
```

```python
df = df.drop(columns=["Boat Type", "Type"])
df.loc[:, categorical_cols.drop(["Material", "id"])] = df_cpy[categorical_cols].groupby(["id"]).max(numeric_only=True)
df
```

```python
df = pd.get_dummies(data = df, columns=['Material'])
df.head(5)
```

```python
df['Manufacturer'] = pd.factorize(df['Manufacturer'])[0]
df.head(5)
```

```python
df['Country'] = pd.factorize(df['Country'])[0]
df.head(5)
```

#### 5.3 удаление несущественных либо избыточных признаков

```python
df = df.drop(columns=["Price", "Location", "currency", "id", "Number of views last 7 days"])
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
target = "price_EUR"
y = df[target]
X = df.drop(columns=[target])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=False)
```

```python
lr = LinearRegression() 
lr.fit(X_train, y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
mean_absolute_error(y_train, lr.predict(X_train))
```

```python
y.describe()
```

```python
r2_score(y_test, y_predict)
```

```python
r2_score(y_train, lr.predict(X_train))
```
