```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

### 1. Загрузите датасет Olivetti faces

```python
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
faces
```

### 2. Познакомьтесь с описанием и структурой датасета. Описание можно найти в документации.

```python
print(faces.DESCR)
```

### 3. Выведите информацию о количественных параметрах датасета

```python
number_of_photos, h, w = faces.images.shape
number_of_targets = np.unique(faces.target).shape[0] # что-то не так
print('Количество фотографий: {}\nКоличество классов: {}'.format(number_of_photos, number_of_targets))
print('Размер фото: {}x{}\nРазмер датасета data: {}\n'.format(w, h, faces.data.shape))
```

### 4.Выведите несколько изображений на экран используя инструментарий библиотеки matplotlib

```python
f= plt.figure(figsize=(10,5))
f.set(facecolor='grey')
plt.suptitle('Фотографии лиц людей',y = 1.1,fontsize=20)
n_row = 5
n_col = 10
for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(faces.images[i], cmap=plt.cm.gray)
        plt.axis('off')
        plt.title(f'Фото: {i}' )
f.subplots_adjust(hspace=0.6,wspace=0.6)
plt.show()
```

### 5. Разделите эти данные на тестовую и обучающую выборки

```python
X = faces.data
y = faces.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=0)
```

### 6. Постройте модель классификатора метода опорных векторов для идентификации человека по изображению

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
(y_test!=y_pred).sum()
```

```python
check = pd.DataFrame({
    "y_test": y_test,
    "y_pred": y_pred,
})

check.head(10)
```

```python
check[check['y_test']!=check['y_pred']]
```

```python
f= plt.figure(figsize=(check[check['y_test']!=check['y_pred']].shape[0],2))
f.set(facecolor='grey')
indexes = check[check['y_test']!=check['y_pred']].index
plt.suptitle('Фотографии лиц людей, которые были определены неверно',y = 1.1,fontsize=20)
n_row = 1
n_col = check[check['y_test']!=check['y_pred']].shape[0]
for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(X_test[indexes[i]].reshape(64, 64), cmap=plt.cm.gray)
        plt.axis('off')
        ind = check[check['y_test']!=check['y_pred']].loc[indexes[i]]
        real = ind["y_test"]
        predict = ind["y_pred"]
        plt.title(f"Real: {real}\nPred: {predict}" )
f.subplots_adjust(hspace=0.6,wspace=0.6)
plt.show()
```

### 7. Оцените качество модели на тестовой выборке с помощью следующих метрик:

```python
from sklearn import metrics
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

#### 7.1. достоверность предсказания (accuracy)

```python
metrics.accuracy_score(y_pred,y_test)
```

#### 7.2. точность (precision)

```python
# Отношение правильно определённых положительных ответов ко всем положительным ответам (которые дала модель)
```

```python
metrics.precision_score(y_pred,y_test, average=None, zero_division = 1)
```

```python
np.average(metrics.precision_score(y_pred,y_test, average=None, zero_division = 1))
```

#### 7.3. полнота (recall);


```python
# Отношение правильно определённых положительных ответов ко всем реальным положительным ответам 
```

```python
metrics.recall_score(y_pred,y_test, average=None, zero_division = 1)
```

```python
np.average(metrics.recall_score(y_pred,y_test, average=None, zero_division = 1))
```

#### 7.4. f1

```python
metrics.f1_score(y_pred,y_test, average=None)
```

```python
np.average(metrics.f1_score(y_pred,y_test, average=None))
```

#### 7.5 Confusion_matrix

```python
from sklearn.metrics import confusion_matrix
conf_mtrx = confusion_matrix(y_test, y_pred)
conf_mtrx
```

### 8. Постройте кривую обучения - график зависимости тестовой и обучающей эффективности от размера обучающей выборки.

```python
training_scores = []
test_scores = []
num = []

for i in np.linspace(20, len(X_train), 20):
    clf = SVC()
    X = X_train[:int(i)]
    y = y_train[:int(i)]
    clf.fit(X, y)
    num.append(int(i))
    training_scores.append(clf.score(X_train, y_train))
    test_scores.append(clf.score(X_test, y_test))

plt.figure(figsize=(10, 5))
plt.plot(num, training_scores, label="training scores")
plt.plot(num, test_scores, label="test scores")
plt.legend();
```
