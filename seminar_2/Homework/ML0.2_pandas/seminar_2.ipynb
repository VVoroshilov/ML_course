## 6. Переименовать все названия колонок на русском языке

```python
# Импортируем библиотеку для автомазиации перевода
from googletrans import Translator
translator = Translator(service_urls=['translate.googleapis.com'])
```

```python
# Функция для перевода
def translate_en_ru(col_en):
    global translator
    return translator.translate(col_en, src="en", dest="ru").text
```

```python
# Перевод имён столбцов
col_names_ru = [translate_en_ru(col_en).capitalize() for col_en in df.columns]
col_names_ru
```

```python
# Создадим словарь для соотнесения английских названий с переводом для метода .rename()
dict_en_ru = dict(zip(list(df.columns), col_names_ru))
dict_en_ru
```

```python
df = df.rename(columns=dict_en_ru)
df
```
