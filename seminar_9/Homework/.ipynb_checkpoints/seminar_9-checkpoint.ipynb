```python
df = pd.DataFrame({"data": news.data, "target": news.target})
df
```

### 5. Разделите эти данные на тестовую и обучающую выборки

```python
X = df["data"]
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=True)
```

```python
pd.DataFrame(X_train, y_train).head(5)
```

### 6. Постройте модель наивного байесовского для классификации текстов

```python
vectorizer = CountVectorizer()
X_count_train = vectorizer.fit_transform(X_train)
```

```python
X_count_train = X_count_train.toarray()
```

```python
X_count_train.shape
```

```python
vectorizer.get_feature_names()
```

```python
X_count_test = vectorizer.transform(X_test)
```

```python
mult_nb = MultinomialNB()
mult_nb.fit(X_count_train, y_train)
```

```python
y_pred = mult_nb.predict(X_count_test)
```

```python
check = pd.DataFrame({
    "y_test": y_test,
    "y_pred": y_pred,
})
check.head(10)
```

### 7. Оцените качество модели на тестовой выборке с помощью следующих метрик:

#### 7.1 достоверность предсказания (accuracy)

```python
accuracy_score(y_test, y_pred)
```

#### 7.2 точность (precision)

```python
precision_score(y_test, y_pred, average=None)
```

#### 7.3 полнота (recall)

```python
recall_score(y_test, y_pred, average=None)
```

#### Confusion matrix

```python
from sklearn.metrics import confusion_matrix
conf_mtrx = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

### 8. Постройте кривую обучения - график зависимости тестовой и обучающей эффективности от размера обучающей выборки. Сделайте вывод о применимости модели.

```python
vectorizer_graph = CountVectorizer(min_df=0.01, max_df=0.05)
X_count_train_graph = vectorizer_graph.fit_transform(X_train)
X_count_train_graph = X_count_train_graph.toarray()
X_count_test_graph = vectorizer_graph.transform(X_test)
```

```python
X_count_test_graph.shape
```

```python
training_scores = []
test_scores = []
num = []

for i in np.linspace(20, len(X_count_train_graph), 20):
    mlt_nb = MultinomialNB()
    X = X_count_train_graph[:int(i)]
    y = y_train.iloc[:int(i)]
    mlt_nb.fit(X, y)
    num.append(int(i))
    training_scores.append(mlt_nb.score(X_count_train_graph, y_train))
    test_scores.append(mlt_nb.score(X_count_test_graph, y_test))

plt.figure(figsize=(10, 5))
plt.plot(num, training_scores, label="training scores")
plt.plot(num, test_scores, label="test scores")
plt.legend();
```
