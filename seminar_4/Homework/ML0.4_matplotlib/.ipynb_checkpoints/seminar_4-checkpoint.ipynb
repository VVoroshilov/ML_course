## 1. Загрузите датасет для предсказани цены квартиры в зависимости от ряда признаков.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df
```

## 2. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

```python
room_counts = df["Rooms"].value_counts(normalize=True)
```

```python
room_counts
```

```python
room_labels = [f"{int(amt)} rooms" for amt in room_counts.keys()]
```

```python
def room_autopct(pct):
    return ('%.2f%%' % pct) if pct > 5 else ''
```

```python
room_counts = df["Rooms"].value_counts()
pie_plot = plt.pie(x = room_counts, autopct=room_autopct, labels = room_labels, explode= [0.15] + [0 for _ in range(1, room_counts.shape[0])])
plt.legend(title="Rooms amount", loc="center right", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Rooms amount")
```

## 3. Постройте гистограмму целевой переменной Price. Оцените визуально, по какой цене продаётся наибольшее количество квартир.

```python
price_counts = df["Price"].value_counts(bins = 10, normalize=True)
price_counts
```

```python
price_labels = [f"{str(amt)} price range" for amt in price_counts.keys()]
```

```python
def price_autopct(pct):
    return ('%.2f%%' % pct) if pct > 5 else ''
```

```python
pie_plot = plt.pie(x = price_counts, autopct=price_autopct, labels = None, explode= [0.15] + [0 for _ in range(1, price_counts.shape[0])])
plt.legend(title="Price range amount", loc="center right", labels = price_labels, bbox_to_anchor=(1, 0, 1, 1))
plt.title("Price amount")
```

```python
plt.hist(df["Price"])
plt.title("Price amount")
plt.xlabel("Price")
plt.ylabel("Amount")
```

## 4. Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области fiagure(). Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.

```python
f=plt.figure()
ax=f.subplots(2,2)

ax[0,0].scatter(df['Rooms'], df['Price'])
ax[0,1].scatter(df['Square'], df['Price'])
ax[1,0].scatter(df['HouseFloor'], df['Price'])
ax[1,1].scatter(df['HouseYear'], df['Price'])

f.set(facecolor = 'grey')
f.subplots_adjust(hspace=0.5,wspace=0.3)

ax[0,0].set(title = 'Rooms', xticks=[], yticks=[])
ax[0,1].set(title = 'Square', xticks=[], yticks=[])
ax[1,0].set(title = 'HouseFloor', xticks=[], yticks=[])
ax[1,1].set(title = 'HouseYear', xticks=[], yticks=[])
plt.suptitle('Рассеивание в зависимости от Price',y = 1.1,fontsize=20)
plt.show()
```

## 7. При помощи сетки графиков PairGrid визуализируйте попарные отношения признаков Rooms, Square, HouseFloor, HouseYear, Price следующим образом: на диагонали - гистограммы, под диагональю - ядерные оценки плотности, над диагональю - диаграммы рассеяния. По результатам визуализации сделайте выводы (аналогичные п.2,3,4).

```python
df_grid = df.copy()[["Rooms", "Square", "HouseFloor", "HouseYear", "Price"]]
df_grid
```

```python
g = sns.PairGrid(df_grid[:5000])
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
plt.show()
```
